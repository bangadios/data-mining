**Membaca data training dan analisis data**

```
# library untuk baca excel
import pandas as pd

# membaca data
data_training = pd.read_csv(train_path)

# print nama kolom dataset
print(data_training.columns.to_list())

# print data di 1 kolom atribut
print(data_training['diagnosis'])

# cetak ukuran data
print(data_training.shape)

# coba cetak 5 data pertama
data_training.head()
```

**Persiapan Data**

```
# pemisahan antara data hasil diagnosis dan kolom atribut
# array berisi nilai
hasil_train = data_training['diagnosis'].to_numpy()
atr_train = data_training[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']].to_numpy()

# library untuk normalisasi data
from sklearn.preprocessing import StandardScaler

# normalisasi
atr_train = StandardScaler().fit_transform(atr_train)
```

**Membuat model**

```
# library untuk membuat model
from sklearn.model_selection import train_test_split

# membagi data jadi training set, dan test set
# - random state = 0  -> membagi secara acak data mana yang dipakai untuk test dan train
# - test_size = 0.20  -> mengatur 20% data sebagai test set
# akan dibagi menjadi 2 jenis seperti konsep klasifikasi (x, y)
X_train, X_test, y_train, y_test = train_test_split(atr_train, hasil_train, test_size = 0.2, random_state=None) # X untuk atribut selain hasil, y adalah hasil
```

**Membuat model classifier dengan Dec Tree**

```
# library untuk decission tree
from sklearn.tree import DecisionTreeClassifier

# fit -> untuk melakukan train pada decission tree
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

**Menguji model classifier yang dibuat ke testing set (validation set)**

```
# melakukan prediksi dari test data berdasarkan train data
y_predict = classifier.predict(X_test)

print(y_predict)
```

**Mengetahui Akurasi testing set (validation set)**

```
# library untuk melakukan uji akurasi
from sklearn.metrics import classification_report, confusion_matrix

tn, fp, fn, tp = confusion_matrix(y_test, y_predict).ravel()
accuracy = (tn + tp) / (tn+fp+fn+tp)
sensitivity = tp / (tp + fn)
specificity = tn / (fp + tn)
print("Akurasi : " + str(accuracy))
print("Sensitivity : " + str(sensitivity))
print("Specificity : " + str(specificity))
```

**Klasifikasi data test dengan model**

```
# membaca data
data_test = pd.read_csv(test_path)

# pemisahan kolom atribut data test
atr_test = data_test[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']].to_numpy()

# normalisasi data test, jika tidak hasil akhir akan sama semua
atr_test = StandardScaler().fit_transform(atr_test)

# dilakukan klasifikasi lagi
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)      # dari data_train.csv

# melakukan prediksi dari data baru
y_predict_test = classifier.predict(atr_test)
print(y_predict_test)
```

**Membuat file excel baru**

```
# membuat data frame baru
y_df = pd.DataFrame(y_predict_test)

ID = data_test.iloc[:, 0]
ID = pd.DataFrame(ID)

hasil = pd.concat([ID, y_df], axis=1)
hasil.to_csv('hasil_klasifikasi.csv')
```
