**Membaca data training dan analisis data**

```
# library untuk baca excel
import pandas as pd
import numpy as np

# membaca data
data_training = pd.read_csv(train_path)
data_testing = pd.read_csv(test_path)
```

**Persiapan data dan preprocessing**

```
# memisahkan antara hasil dan atribut
# array berisi nilai
hasil_train = data_training['diagnosis'].to_numpy()
atr_train = data_training[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']].to_numpy()
atr_test = data_testing[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']].to_numpy()

# library untuk normalisasi data
# range data dirubah menjadi antara 0 dan 1
from sklearn.preprocessing import MinMaxScaler

# normalisasi
scaler = MinMaxScaler(feature_range=(0,1))
atr_train = scaler.fit_transform(atr_train)
atr_test = scaler.fit_transform(atr_test)

print(atr_train)
print(atr_test)
```

**Membuat model, membuat prediksi, dan melakukan klasifikasi dengan model terbaik**

```
# library untuk membuat model
from sklearn.model_selection import KFold

# library untuk decission tree
from sklearn.tree import DecisionTreeClassifier

# nilai klasik untuk random_state = 0 dan 42
# n_splits menentukan banyak iterasi/epoch
kfold = KFold(n_splits = 10, random_state = None, shuffle=False)
best_dTree = DecisionTreeClassifier()
score = 0
nomor_model = 0
y_pred_test = []

i = 1
for train_index, test_index in kfold.split(atr_train):
  
  # print(f"Epoch {i} \n")
  # print(f"Train Index : {train_index} \n")
  # print(f"Test Index  : {test_index}")

  X_train, X_test, y_train, y_test = atr_train[train_index], atr_train[test_index], hasil_train[train_index], hasil_train[test_index]
  
  # fit -> untuk melakukan train pada decission tree
  best_dTree.fit(X_train, y_train)
  y_predict = best_dTree.predict(X_test)

  # perhitungan akurasi model
  akurasi = best_dTree.score(X_test, y_test)

  print(f'Model ke {i}')
  print(f'Akurasi : {akurasi}')
  print(f'Hasil Prediksi (dari data training): {y_predict}\n')

  # memilih model dengan akurasi terbaik
  if score < akurasi:
    score = akurasi
    nomor_model = i

    # melakukan klasifikasi data_testing.csv menggunakan model dengan akurasi terbaik
    y_pred_test = best_dTree.predict(atr_test)
  elif score == akurasi:
    pass
  elif score > akurasi:
    pass

  i+=1
```

**Cetak hasil klasifikasi data testing menggunakan model terbaik**

```
print(f'Klasifikasi menggunakan nomor model ke - {nomor_model}')
print(f'Dengan akurasi : {score}')
print(y_pred_test)
```

**Membuat file excel baru**

```
# membuat data frame baru
y_df = pd.DataFrame(y_pred_test)

ID = data_test.iloc[:, 0]
ID = pd.DataFrame(ID)

hasil = pd.concat([ID, y_df], axis=1)
hasil.to_csv('hasil_klasifikasi.csv')
```
